```python
#importing libraries

import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
#send request,give url.

url = 'https://www.imdb.com/list/ls068082370/'

page=requests.get(url)

#creating soup using the library BeautifulSoup

soup = BeautifulSoup(page.text,'html.parser')
```

```python
#index numbers

id_num = []

id_elem = soup.find_all(class_='lister-item-index unbold text-primary')

for item in id_elem:
    id_num.append(item.text)
    
id_num
```
